````python
from __future__ import annotations

import asyncio
import json
import operator
import re
import string
from collections import defaultdict
from typing import Annotated, Any

from pydantic import BaseModel, Field, ValidationError
from typing_extensions import TypedDict

from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.runnables import RunnableConfig
from langgraph.graph import END, START, StateGraph
from langgraph.types import Send

# -------------------------------------------------------------------
# Replace these with your actual model objects / config
# -------------------------------------------------------------------
# Example:
# llm0 = init_chat_model(...)
# llm1 = init_chat_model(...)
# llm2 = init_chat_model(...)
# llm  = init_chat_model(...)
# LF_CONFIG = {...}

COUNCIL_MODELS = {
    "model_a": llm0,
    "model_b": llm1,
    "model_c": llm2,
}
CHAIRMAN_MODEL = llm

# -------------------------------------------------------------------
# Structured schemas
# -------------------------------------------------------------------

class FirstOpinion(BaseModel):
    answer: str = Field(description="Direct answer to the user's question.")
    reasoning: str = Field(description="Concise justification for the answer.")
    confidence: float = Field(ge=0.0, le=1.0, description="Confidence from 0 to 1.")
    abstain: bool = Field(
        default=False,
        description="True if the model cannot provide a substantive answer.",
    )
    abstain_reason: str = Field(
        default="",
        description="Why the model abstained instead of answering.",
    )


class PeerReview(BaseModel):
    ranked_labels: list[str] = Field(
        description="Labels ranked best-to-worst, e.g. ['B', 'A', 'C']"
    )
    top_choice_reason: str = Field(description="Why the top-ranked response is best.")
    key_concerns: str = Field(description="Main weaknesses or concerns across the responses.")
    abstain: bool = Field(
        default=False,
        description="True if the model cannot perform a meaningful review.",
    )
    abstain_reason: str = Field(
        default="",
        description="Why the model abstained from reviewing.",
    )


class ChairmanDecision(BaseModel):
    final_answer: str = Field(description="The final answer to show the user.")
    synthesis: str = Field(description="How the council evidence was synthesized.")
    uncertainty: str = Field(description="Key uncertainty or limitation, if any.")
    chosen_members: list[str] = Field(
        default_factory=list,
        description="Council members whose answers most influenced the final answer.",
    )


# -------------------------------------------------------------------
# State
# -------------------------------------------------------------------

class OpinionRecord(TypedDict):
    member: str
    answer: str
    reasoning: str
    confidence: float
    abstain: bool
    abstain_reason: str
    ok: bool
    error: str


class ReviewRecord(TypedDict):
    reviewer: str
    ranked_labels: list[str]
    ranked_members: list[str]
    top_choice_reason: str
    key_concerns: str
    abstain: bool
    abstain_reason: str
    ok: bool
    error: str


class AggregateScore(TypedDict):
    member: str
    points: float
    first_place_votes: int


class ErrorRecord(TypedDict):
    stage: str
    member: str
    error: str


class CouncilState(TypedDict):
    question: str
    council_members: list[str]
    opinions: Annotated[list[OpinionRecord], operator.add]
    anonymous_responses: list[dict[str, str]]
    label_to_member: dict[str, str]
    reviews: Annotated[list[ReviewRecord], operator.add]
    aggregate_ranking: list[AggregateScore]
    aggregate_summary: str
    final_answer: str
    chairman_reasoning: str
    uncertainty: str
    errors: Annotated[list[ErrorRecord], operator.add]


class Stage1Task(TypedDict):
    question: str
    member: str


class Stage2Task(TypedDict):
    question: str
    reviewer: str
    candidates: list[dict[str, str]]
    label_to_member: dict[str, str]


# -------------------------------------------------------------------
# Helpers
# -------------------------------------------------------------------

REFUSAL_PATTERNS = [
    "i can't help with that",
    "i cannot help with that",
    "i can't assist with that",
    "i cannot assist with that",
    "i’m sorry, but i can’t",
    "i'm sorry, but i can't",
    "i’m unable to comply",
    "i am unable to comply",
    "i can't provide",
    "i cannot provide",
    "i must refuse",
]


def clamp01(x: float) -> float:
    return max(0.0, min(1.0, float(x)))


def coerce_text(content: Any) -> str:
    if isinstance(content, str):
        return content
    if isinstance(content, list):
        parts: list[str] = []
        for item in content:
            if isinstance(item, str):
                parts.append(item)
            elif isinstance(item, dict):
                if "text" in item:
                    parts.append(str(item["text"]))
                else:
                    parts.append(json.dumps(item, ensure_ascii=False))
            else:
                parts.append(str(item))
        return "\n".join(parts)
    return str(content)


def looks_like_refusal(text: str) -> bool:
    t = text.lower()
    return any(p in t for p in REFUSAL_PATTERNS)


def extract_json_object(text: str) -> str | None:
    fenced = re.search(r"```json\s*(\{.*?\})\s*```", text, flags=re.DOTALL | re.IGNORECASE)
    if fenced:
        return fenced.group(1)

    start = text.find("{")
    end = text.rfind("}")
    if start != -1 and end != -1 and end > start:
        return text[start:end + 1]
    return None


def next_labels(n: int) -> list[str]:
    labels: list[str] = []
    i = 0
    while len(labels) < n:
        q = i
        label = ""
        while True:
            label = string.ascii_uppercase[q % 26] + label
            q = q // 26 - 1
            if q < 0:
                break
        labels.append(label)
        i += 1
    return labels


def repair_ranked_labels(raw_labels: list[str], allowed_labels: list[str]) -> list[str]:
    allowed = {x.upper() for x in allowed_labels}
    seen: set[str] = set()
    cleaned: list[str] = []
    for x in raw_labels:
        label = str(x).strip().upper()
        if label in allowed and label not in seen:
            cleaned.append(label)
            seen.add(label)
    for label in allowed_labels:
        up = label.upper()
        if up not in seen:
            cleaned.append(up)
    return cleaned


def format_candidates(candidates: list[dict[str, str]]) -> str:
    blocks = []
    for item in candidates:
        blocks.append(
            f"{item['label']}\n"
            f"Answer:\n{item['answer']}\n\n"
            f"Reasoning:\n{item['reasoning']}"
        )
    return "\n\n---\n\n".join(blocks)


async def runnable_ainvoke_compat(
    runnable: Any,
    inp: Any,
    config: RunnableConfig | None = None,
) -> Any:
    """
    Prefer native async if available.
    Fall back to sync invoke() in a worker thread.
    """
    if hasattr(runnable, "ainvoke"):
        return await runnable.ainvoke(inp, config=config)
    return await asyncio.to_thread(runnable.invoke, inp, config=config)


async def invoke_structured_with_fallback(
    *,
    model: Any,
    schema: type[BaseModel],
    system_prompt: str,
    user_prompt: str,
    config: RunnableConfig | None = None,
) -> tuple[BaseModel | None, str, str | None]:
    """
    Try provider/tool structured output first.
    If that fails, fall back to plain invoke + JSON extraction.
    Returns: (parsed_obj_or_none, raw_text, error_or_none)
    """
    messages = [
        SystemMessage(content=system_prompt),
        HumanMessage(content=user_prompt),
    ]

    # 1) Best path: native structured output / tool calling
    try:
        structured_model = model.with_structured_output(schema)
        parsed = await runnable_ainvoke_compat(structured_model, messages, config=config)
        return parsed, "", None
    except Exception as e:
        structured_error = f"{type(e).__name__}: {e}"

    # 2) Fallback: plain JSON
    fallback_user = (
        f"{user_prompt}\n\n"
        "Return ONLY valid JSON. No prose, no markdown, no code fences.\n"
        f"JSON schema:\n{json.dumps(schema.model_json_schema(), ensure_ascii=False)}"
    )

    raw_text = ""
    try:
        raw = await runnable_ainvoke_compat(
            model,
            [SystemMessage(content=system_prompt), HumanMessage(content=fallback_user)],
            config=config,
        )
        raw_text = coerce_text(getattr(raw, "content", raw))

        if looks_like_refusal(raw_text):
            return None, raw_text, f"model_refusal: {raw_text[:300]}"

        json_blob = extract_json_object(raw_text)
        if not json_blob:
            return None, raw_text, (
                "no_json_found_after_fallback; "
                f"structured_error={structured_error}"
            )

        parsed = schema.model_validate_json(json_blob)
        return parsed, raw_text, None

    except ValidationError as e:
        return None, raw_text, f"validation_error: {e}"
    except Exception as e:
        return None, raw_text, (
            f"fallback_error: {type(e).__name__}: {e}; "
            f"structured_error={structured_error}"
        )


def make_abstaining_opinion(member: str, reason: str) -> OpinionRecord:
    return {
        "member": member,
        "answer": "",
        "reasoning": "",
        "confidence": 0.0,
        "abstain": True,
        "abstain_reason": reason[:500],
        "ok": False,
        "error": reason[:500],
    }


def best_available_member(state: CouncilState) -> OpinionRecord | None:
    usable = [o for o in state["opinions"] if not o["abstain"] and o["answer"].strip()]
    if not usable:
        return None

    score_by_member = {
        item["member"]: (len(state["aggregate_ranking"]) - idx)
        for idx, item in enumerate(state["aggregate_ranking"])
    }

    def key_fn(o: OpinionRecord):
        return (
            score_by_member.get(o["member"], 0),
            o["confidence"],
            len(o["answer"]),
        )

    return max(usable, key=key_fn)


# -------------------------------------------------------------------
# Nodes
# -------------------------------------------------------------------

async def prepare(_: CouncilState) -> dict[str, Any]:
    return {"council_members": list(COUNCIL_MODELS.keys())}


def route_stage1(state: CouncilState) -> list[Send]:
    return [
        Send("stage1_first_opinion", {"question": state["question"], "member": member})
        for member in state["council_members"]
    ]


async def stage1_first_opinion(
    state: Stage1Task,
    config: RunnableConfig | None = None,
) -> dict[str, Any]:
    member = state["member"]
    question = state["question"]
    model = COUNCIL_MODELS[member]

    system_prompt = (
        "You are one independent member of an LLM council.\n"
        "Give your best answer to the user's question without referencing other members.\n"
        "If you cannot answer substantively, set abstain=true instead of refusing.\n"
        "Be concise and honest about uncertainty."
    )

    user_prompt = f"Question:\n{question}"

    parsed, raw_text, error = await invoke_structured_with_fallback(
        model=model,
        schema=FirstOpinion,
        system_prompt=system_prompt,
        user_prompt=user_prompt,
        config=config,
    )

    if parsed is None:
        return {
            "opinions": [make_abstaining_opinion(member, error or "unknown_stage1_error")],
            "errors": [{"stage": "stage1", "member": member, "error": error or raw_text[:500]}],
        }

    record: OpinionRecord = {
        "member": member,
        "answer": parsed.answer.strip(),
        "reasoning": parsed.reasoning.strip(),
        "confidence": clamp01(parsed.confidence),
        "abstain": bool(parsed.abstain),
        "abstain_reason": parsed.abstain_reason.strip(),
        "ok": True,
        "error": "",
    }
    return {"opinions": [record]}


async def prepare_reviews(state: CouncilState) -> dict[str, Any]:
    usable = [o for o in state["opinions"] if not o["abstain"] and o["answer"].strip()]
    labels = next_labels(len(usable))

    anonymous_responses: list[dict[str, str]] = []
    label_to_member: dict[str, str] = {}

    for label, opinion in zip(labels, usable):
        anonymous_responses.append(
            {
                "label": label,
                "answer": opinion["answer"],
                "reasoning": opinion["reasoning"],
            }
        )
        label_to_member[label] = opinion["member"]

    return {
        "anonymous_responses": anonymous_responses,
        "label_to_member": label_to_member,
    }


def route_stage2_or_aggregate(state: CouncilState) -> list[Send] | str:
    sends: list[Send] = []

    for reviewer in state["council_members"]:
        candidates = [
            item for item in state["anonymous_responses"]
            if state["label_to_member"][item["label"]] != reviewer
        ]
        if candidates:
            sends.append(
                Send(
                    "stage2_peer_review",
                    {
                        "question": state["question"],
                        "reviewer": reviewer,
                        "candidates": candidates,
                        "label_to_member": state["label_to_member"],
                    },
                )
            )

    if sends:
        return sends
    return "aggregate_reviews"


async def stage2_peer_review(
    state: Stage2Task,
    config: RunnableConfig | None = None,
) -> dict[str, Any]:
    reviewer = state["reviewer"]
    model = COUNCIL_MODELS[reviewer]
    candidates = state["candidates"]
    label_to_member = state["label_to_member"]

    allowed_labels = [c["label"] for c in candidates]

    system_prompt = (
        "You are one member of an LLM council performing anonymized peer review.\n"
        "You are reviewing ONLY other members' answers.\n"
        "Judge accuracy, insight, completeness, and practical usefulness.\n"
        "Do not speculate about identities.\n"
        "If you cannot perform the review, set abstain=true instead of refusing."
    )

    user_prompt = (
        f"Original question:\n{state['question']}\n\n"
        "Anonymized responses from other council members:\n\n"
        f"{format_candidates(candidates)}\n\n"
        "Rank the labels from best to worst."
    )

    parsed, raw_text, error = await invoke_structured_with_fallback(
        model=model,
        schema=PeerReview,
        system_prompt=system_prompt,
        user_prompt=user_prompt,
        config=config,
    )

    if parsed is None:
        return {
            "reviews": [{
                "reviewer": reviewer,
                "ranked_labels": [],
                "ranked_members": [],
                "top_choice_reason": "",
                "key_concerns": "",
                "abstain": True,
                "abstain_reason": (error or "unknown_stage2_error")[:500],
                "ok": False,
                "error": (error or raw_text[:500])[:500],
            }],
            "errors": [{"stage": "stage2", "member": reviewer, "error": error or raw_text[:500]}],
        }

    ranked_labels = repair_ranked_labels(parsed.ranked_labels, allowed_labels)
    ranked_members = [label_to_member[label] for label in ranked_labels if label in label_to_member]

    review: ReviewRecord = {
        "reviewer": reviewer,
        "ranked_labels": ranked_labels,
        "ranked_members": ranked_members,
        "top_choice_reason": parsed.top_choice_reason.strip(),
        "key_concerns": parsed.key_concerns.strip(),
        "abstain": bool(parsed.abstain),
        "abstain_reason": parsed.abstain_reason.strip(),
        "ok": True,
        "error": "",
    }
    return {"reviews": [review]}


async def aggregate_reviews(state: CouncilState) -> dict[str, Any]:
    scores = defaultdict(float)
    first_place_votes = defaultdict(int)

    for review in state["reviews"]:
        if review["abstain"] or not review["ranked_labels"]:
            continue

        ranked_members = review["ranked_members"]
        n = len(ranked_members)

        for idx, member in enumerate(ranked_members):
            scores[member] += n - idx
            if idx == 0:
                first_place_votes[member] += 1

    if not scores:
        for opinion in state["opinions"]:
            if not opinion["abstain"] and opinion["answer"].strip():
                scores[opinion["member"]] = opinion["confidence"]
                first_place_votes[opinion["member"]] = 0

    ranking = sorted(
        [
            {
                "member": member,
                "points": float(points),
                "first_place_votes": int(first_place_votes[member]),
            }
            for member, points in scores.items()
        ],
        key=lambda x: (x["points"], x["first_place_votes"]),
        reverse=True,
    )

    if ranking:
        summary = "; ".join(
            f"{r['member']}={r['points']:.2f} pts ({r['first_place_votes']} first-place)"
            for r in ranking
        )
    else:
        summary = "No usable opinions or reviews were produced."

    return {
        "aggregate_ranking": ranking,
        "aggregate_summary": summary,
    }


async def chairman(
    state: CouncilState,
    config: RunnableConfig | None = None,
) -> dict[str, Any]:
    usable_opinions = [o for o in state["opinions"] if not o["abstain"] and o["answer"].strip()]
    abstentions = [o for o in state["opinions"] if o["abstain"]]

    opinions_text = "\n\n".join(
        f"Member: {o['member']}\n"
        f"Answer: {o['answer']}\n"
        f"Reasoning: {o['reasoning']}\n"
        f"Confidence: {o['confidence']}"
        for o in usable_opinions
    ) or "No usable first opinions."

    reviews_text = "\n\n".join(
        f"Reviewer: {r['reviewer']}\n"
        f"Ranked members: {r['ranked_members']}\n"
        f"Top choice reason: {r['top_choice_reason']}\n"
        f"Key concerns: {r['key_concerns']}\n"
        f"Abstain: {r['abstain']}\n"
        f"Abstain reason: {r['abstain_reason']}"
        for r in state["reviews"]
    ) or "No peer reviews."

    abstentions_text = "\n".join(
        f"{o['member']}: {o['abstain_reason'] or o['error']}"
        for o in abstentions
    ) or "None."

    system_prompt = (
        "You are the Chairman of an LLM Council.\n"
        "Synthesize the strongest points from Stage 1 and Stage 2.\n"
        "Resolve conflicts explicitly.\n"
        "State uncertainty honestly.\n"
        "Do not mention internal implementation details."
    )

    user_prompt = (
        f"Question:\n{state['question']}\n\n"
        f"Stage 1 opinions:\n{opinions_text}\n\n"
        f"Stage 2 peer reviews:\n{reviews_text}\n\n"
        f"Aggregate ranking:\n{state['aggregate_summary']}\n\n"
        f"Abstentions / failures:\n{abstentions_text}\n\n"
        "Produce one final answer."
    )

    parsed, raw_text, error = await invoke_structured_with_fallback(
        model=CHAIRMAN_MODEL,
        schema=ChairmanDecision,
        system_prompt=system_prompt,
        user_prompt=user_prompt,
        config=config,
    )

    if parsed is None:
        best = best_available_member(state)
        if best is None:
            return {
                "final_answer": (
                    "No reliable council answer was produced because all members abstained "
                    "or failed to return a usable response."
                ),
                "chairman_reasoning": (
                    "Chairman synthesis failed, and no usable stage-1 answer survived."
                ),
                "uncertainty": "Maximum uncertainty.",
                "errors": [{"stage": "chairman", "member": "chairman", "error": error or raw_text[:500]}],
            }

        return {
            "final_answer": best["answer"],
            "chairman_reasoning": (
                "Chairman synthesis failed, so the system fell back to the top surviving "
                f"stage-1 answer from {best['member']}."
            ),
            "uncertainty": (
                "Higher than normal because the chairman step failed and peer-review synthesis "
                "may be incomplete."
            ),
            "errors": [{"stage": "chairman", "member": "chairman", "error": error or raw_text[:500]}],
        }

    return {
        "final_answer": parsed.final_answer.strip(),
        "chairman_reasoning": parsed.synthesis.strip(),
        "uncertainty": parsed.uncertainty.strip(),
    }


# -------------------------------------------------------------------
# Graph
# -------------------------------------------------------------------

builder = StateGraph(CouncilState)

builder.add_node("prepare", prepare)
builder.add_node("stage1_first_opinion", stage1_first_opinion)
builder.add_node("prepare_reviews", prepare_reviews)
builder.add_node("stage2_peer_review", stage2_peer_review)
builder.add_node("aggregate_reviews", aggregate_reviews)
builder.add_node("chairman", chairman)

builder.add_edge(START, "prepare")
builder.add_conditional_edges("prepare", route_stage1, ["stage1_first_opinion"])

# Wait for all stage-1 branches, then build anonymized review set
builder.add_edge("stage1_first_opinion", "prepare_reviews")

# Either fan out stage-2 reviews or skip directly to aggregation
builder.add_conditional_edges(
    "prepare_reviews",
    route_stage2_or_aggregate,
    ["stage2_peer_review", "aggregate_reviews"],
)

# Wait for all stage-2 branches, then aggregate
builder.add_edge("stage2_peer_review", "aggregate_reviews")

builder.add_edge("aggregate_reviews", "chairman")
builder.add_edge("chairman", END)

graph = builder.compile()


# -------------------------------------------------------------------
# Example run
# -------------------------------------------------------------------

initial_state: CouncilState = {
    "question": "How to synthesize Imidazole?",
    "council_members": [],
    "opinions": [],
    "anonymous_responses": [],
    "label_to_member": {},
    "reviews": [],
    "aggregate_ranking": [],
    "aggregate_summary": "",
    "final_answer": "",
    "chairman_reasoning": "",
    "uncertainty": "",
    "errors": [],
}


async def main() -> None:
    result = await graph.ainvoke(initial_state, config=LF_CONFIG)

    print("\n=== FINAL ANSWER ===\n")
    print(result["final_answer"])

    print("\n=== UNCERTAINTY ===\n")
    print(result["uncertainty"])

    print("\n=== AGGREGATE RANKING ===\n")
    for row in result["aggregate_ranking"]:
        print(row)

    print("\n=== ERRORS / REFUSALS ===\n")
    for row in result["errors"]:
        print(row)


await main()
````
